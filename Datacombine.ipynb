```python
import os
import yaml
import pandas as pd

BASE_DIR = r"C:/Users/Karthika/Downloads/data"

symbol_data = {}
all_rows = []  


for month_folder in os.listdir(BASE_DIR):
    month_path = os.path.join(BASE_DIR, month_folder)
    if not os.path.isdir(month_path):
        continue
    
    for filename in os.listdir(month_path):
        if filename.endswith(".yaml"):
            file_path = os.path.join(month_path, filename)

            with open(file_path, "r") as file:
                data = yaml.safe_load(file)

           
            if isinstance(data, dict):
                data = list(data.values())[0]

            for entry in data:
                symbol = entry["Ticker"]

                row = entry.copy()

             
                symbol_data.setdefault(symbol, []).append(row)

                
                all_rows.append(row)


OUTPUT_DIR = r"C:/Users/Karthika/Downloads/output_csv"
os.makedirs(OUTPUT_DIR, exist_ok=True)


combined_csv_path = os.path.join(OUTPUT_DIR, "combined_output.csv")
pd.DataFrame(all_rows).to_csv(combined_csv_path, index=False)
print("Combined CSV created:", combined_csv_path)


for symbol, rows in symbol_data.items():
    df = pd.DataFrame(rows)
    df.to_csv(f"{OUTPUT_DIR}/{symbol}.csv", index=False)

print("Symbol-wise CSV files created successfully!")
```

```python

results = []

for ticker in df["Ticker"].unique():
    ticker_df = df[df["Ticker"] == ticker]

    # 2023 data (for Open)
    df_2023 = ticker_df[ticker_df["Year"] == 2023].sort_values("date")

    # 2024 data (for Close)
    df_2024 = ticker_df[ticker_df["Year"] == 2024].sort_values("date")

    # Skip if missing any year
    if df_2023.empty or df_2024.empty:
        continue

    open_2023 = df_2023.iloc[0]["open"]       
    close_2024 = df_2024.iloc[-1]["close"]     

    return_pct = ((close_2024 - open_2023) / open_2023) * 100

    results.append({
        "Ticker": ticker,
        "Open_2023": round(open_2023, 2),
        "Close_2024": round(close_2024, 2),
        "Return_2023_2024_%": round(return_pct, 2)
    })

result_df = pd.DataFrame(results)
result_df
```

```python
top_5_cumlative = (
    result_df
    .sort_values("Return_2023_2024_%", ascending=False)
    .head(5)
    .reset_index(drop=True)
)

top_5_cumlative
```

```python
results = []


for ticker in df["Ticker"].unique():
    ticker_df = df[df["Ticker"] == ticker].sort_values("Date")


    for i in range(1, len(ticker_df)):
        prev_close = ticker_df.iloc[i - 1]["close"]
        curr_close = ticker_df.iloc[i]["close"]

        daily_return = (curr_close - prev_close) / prev_close

        results.append({
            "Ticker": ticker,
            "Date": ticker_df.iloc[i]["Date"],
            "Prev_Close": round(prev_close, 2),
            "Close": round(curr_close, 2),
            "Daily_Return": round(daily_return, 6)
        })

daily_return_df = pd.DataFrame(results)
daily_return_df


```

```python
daily_return_df["Std_Dev"] = (
    daily_return_df
    .groupby("Ticker")["Daily_Return"]
    .transform("std")
)
daily_return_df
```

```python
top_10_volatility = (
    daily_return_df.groupby("Ticker", as_index=False)["Std_Dev"]
      .mean()
      .sort_values("Std_Dev", ascending=False)
      .head(10)
)
top_10_volatility
```

```python
top_10_volatility.to_csv("top_10_volatility", index=False)
```

```python
stock_prices = (
    df.groupby("Ticker")
      .agg(
          Start_Price=("open", "first"),
          End_Price=("close", "last"),
          Avg_Volume=("volume", "mean")
      )
      .reset_index()
)
stock_prices
```

```python
stock_prices["Return_%"] = (
    (stock_prices["End_Price"] - stock_prices["Start_Price"])
    / stock_prices["Start_Price"]
) * 100
stock_prices
```

```python
top_10_green = stock_prices.sort_values(
    "Return_%", ascending=False
).head(10)
top_10_green
```

```python
top_10_green.to_csv("top_10_green", index=False)
```

```python
bottom_10_red = stock_prices.sort_values(
    "Return_%",ascending=True).head(10)
bottom_10_red 
```

```python
bottom_10_red.to_csv("bottom_10_red",index=False)
```

```python
green_stocks = (stock_prices["Return_%"] > 0).sum()
red_stocks = (stock_prices["Return_%"] <= 0).sum()
green_stocks
red_stocks
```

```python
average_price = (
    stock_prices[["Start_Price", "End_Price"]].mean().mean()
)
average_price
```

```python
average_volume = stock_prices["Avg_Volume"].mean()
average_volume
```

```python
market_summary = pd.DataFrame({
    "Metric": [
        "Green Stocks",
        "Red Stocks",
        "Average Price",
        "Average Volume"
    ],
    "Value": [
        green_stocks,
        red_stocks,
        round(average_price, 2),
        int(average_volume)
    ]
})

market_summary
```

```python
market_summary.to_csv("market_summary",index=False)
```
